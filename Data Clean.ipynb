```python
import pandas as pd
df = pd.read_csv('exercise_data.csv')
df.head() #dirty, integers are shown alongside strings
```

```python
df.describe() #more "clean" since it's only showing integers
```

```python
df['age'].unique()
df['age'] = pd.to_numeric(df['age'], errors = 'coerce') #cleans the data for names
```

```python
df.info() #once again some "dirty" data, it's barely understandable.
```

```python
df['num_hairs'].describe() #cleans the data regarding number of hairs
```

```python
import numpy as np
df['num_hairs'].replace([np.inf, -np.inf], np.nan, inplace = True) #replaces infinite values, making it more "clean"
df.describe()
```

```python
df['name'].replace('99999', np.nan, inplace = True) # this line selects only strings from the data set, even if there are "dirty" numbers in the column of names
df['name'].unique()    #it gest the unique names form the data set to avoid repetition.
```

```python
data_merged_df = pd.read_csv('merge_file (1).csv')
#print(merger_df.shape)
new_data_df = df.merge(data_merged_df, on = 'name') #on value is column that you merge on
new_data_df.head() #print the data merged into one single table
```
